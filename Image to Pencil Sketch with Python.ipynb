<b><h1>Importing Libraries<b><h1>
    
Import in python is similar to #include header_file in C/C++. Python modules can get access to code from another module by importing the file/function using import.

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
from cv2 import imread                                                                                                              
```

 <b><h1>loading the dataset<b><h1>
    
The Load method provides a technique for filling a single DataTable with data, retrieved from an IDataReader instance. This method provides the same functionality

```python
image = cv2.imread('bappa.jpg')
```

<b><h1>To show or display image <b><h1>

```python
plt.figure(figsize=(16,14))
plt.imshow(image)
plt.title("Image")
plt.axis("off")
```

<b><h1>Read the image in RGB format<b><h1>

```python
plt.figure(figsize = (16,14))
convert_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(convert_image)
plt.title("original_image")
plt.axis("off")
```

<b><h1>Converting the original image to grayscale image<b><h1>

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(gray_image, cmap='gray')
plt.title("Grascale_image")
plt.axis("off")
plt.show()
```

<b><h1>Inverting the image<b><h1>

```python
inverted_image = cv2.bitwise_not(gray_image)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(inverted_image, cmap='gray')
plt.title("inverted_image")
plt.axis("off")
plt.show()
```

<b><h1>Blur the image by using Gaussian function<b><h1>

```python
gaus_blur = cv2.GaussianBlur(inverted_image,(111,111),0)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(gaus_blur, cmap='gray')
plt.title("blured_image")
plt.axis("off")
plt.show()
```

<b><h1>Inverting the blured image<b><h1>

```python
invertedblur = cv2.bitwise_not(gaus_blur)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(invertedblur, cmap='gray')
plt.axis("off")
plt.show()
```

<b><h1>Final pencil sketch<b><h1>

```python
sketch = cv2.divide(gray_image,invertedblur,scale=255)
```

```python
plt.figure(figsize=(16,14))
plt.imshow(sketch, cmap='gray')
plt.title("pencil_sketch")
plt.axis("off")
plt.show()
```
